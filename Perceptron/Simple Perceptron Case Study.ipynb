## Possible Weights and Bias:

For __Threshold = 35__
- [1.6, 0.2, -1.5]
- [1.2, 0.2, -0.5]

For __Threshold = 45__
- [1.8, 0.2, -1.5]
- [1.0, 0.4, -1.9]
- [(0.8, 0.9, 1.0), 0.4, -1.8]


```python
import numpy as np
# x1 = int(input('How many hours did the student study? '))
# x2 = int(input("Marks obtained: "))

# implementing a simple perceptron
def perceptron(input1, input2):
    weights = np.array([1.8,0.2])
    bias = -1.5
    x = [input1, input2]
    val = np.dot(weights,x) + bias
    print("value = ", val)
    if val >= 45:
        return "1"
    else:
        return "0"


print("Output: "+ perceptron(35, 67) + " Target: 1")
print("Output: "+ perceptron(12, 75)+ " Target: 0")
print("Output: "+ perceptron(16, 89)+ " Target: 1")
print("Output: "+ perceptron(45, 56)+ " Target: 1")
print("Output: "+ perceptron(10, 90)+ " Target: 0")
```

```python
# validation case
print("Output: "+ perceptron(25, 70) + " Target: 1")
```
